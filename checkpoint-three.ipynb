## Irregular Data

Detect outliers in your dataset and handle them as needed. Use code comments to make notes about your thought process.

```python
# max_model = data.sort_values(by='Model Year', ascending=False).head(10)
# print(max_model)
# The Model year has some futuristic dates that will need to go.

# over_2017 = np.mean(data['Model Year'] > 2017)
# print('Model Year - {}%'.format(round(over_2017 * 100)))
# 2017 is the most recent year in the data set, this shows that 6% of the model year columns is above that line meaning it is innaccurate.

data = data[data['Model Year'] <=2017]

max_model = data.sort_values(by='Model Year', ascending=False).head(10)
print(max_model)
```

```python
q1 = data['Estimated Units'].quantile(.25)
q3 = data['Estimated Units'].quantile(.75)

iqr = q3 - q1

lower_fence = q1 - (1.5 * iqr) 
upper_fence = q3 + (1.5 * iqr)

upper_outliers = data[data['Estimated Units'] > upper_fence]
lower_outliers = data[data['Estimated Units'] < lower_fence]

percentage_above_fence = (len(upper_outliers) / len(data)) * 100
percentage_below_fence = (len(lower_outliers) / len(data)) * 100

print('Lower Outliers %:',percentage_below_fence)
print('Upper Outliers %:',percentage_above_fence)


# I found these outliers in the estimated units but since it doesnt seemthe product of an error im going to leave them in.
```
